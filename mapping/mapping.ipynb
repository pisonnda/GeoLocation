```python
import requests
import json
import sys
import os
import re
import folium

wifilist = []
```

```python
def main():
	count = 0
	filename = "sapporo.txt"
    #filename = sys.argv[1]
	with open(filename, "r") as fp:
			buf = fp.readlines()
	result = buf[1:]

	#Get list of wifi and print to screen
	print("List of wifi")
	for row in result:
		a = re.sub('^ +', '', row)
		a = re.sub(' +', ',', row)
		a = a.split(",")
		SSID = a[1]
		MAC  = a[2]
		RSSI = a[3]
		print("({})\t{}\t{}\t{}".format(count, SSID.ljust(16, " "), MAC, RSSI))
		wifilist.append(a[1:4])
		count += 1
	print("-------------------------------------")

	#Choose Wifi -> Get MAC -> Get Location
	i = 0
	j = len(wifilist)
	leng = len(wifilist)
	for i in range(leng):
		print("Select Wifi1, 2: ", end='')
		n1, n2 = input().split(",")
		n1 = int(n1)
		n2 = int(n2)

		mac1 = str(wifilist[n1][1])
		#print(mac1)
		rssi1 = str(wifilist[n1][2])
		mac2 = str(wifilist[n2][1])
		#print(mac2)
		rssi2 = str(wifilist[n2][2])

		location = getlocation(mac1, rssi1, mac2, rssi2)
		#print("=================Your location:==============")
		try: 
			coordinate = location['location']
			print(coordinate)
		except:
			print("ERROR: Invalid Wifi") 
			continue
		lat = coordinate['lat']
		lng = coordinate['lng']
		mapp = mapping(lat,lng)
		mapp
		#print("---------------------------------------------")
```

```python
#Get location Function
def getlocation(mac1, rssi1, mac2, rssi2):
        #POST URL
        url = "https://www.googleapis.com/geolocation/v1/geolocate?key=AIzaSyCeoRn12KZ3-WXXcVRhSnLVy2jdPF1H1Yw"
        #JSON
        json_data = {
        "considerIp": "false",
        "wifiAccessPoints": [
        {
        "macAddress": mac1,
        "signalStrength": rssi1,
        "signalToNoiseRatio": 0
        },
        {
        "macAddress": mac2,
        "signalStrength": rssi2,
        "signalToNoiseRatio": 0
        }
        ]
        }
        #POST 送信
        #print(json_data)
        response = requests.post(
                url, 
                data = json.dumps(json_data)
                )
        location = response.json()
        #print(type(location))
        return location
```

```python
def mapping(lat, long):
    latitude = lat
    longtude = long
    name = "location" 
    map = folium.Map(location=[latitude, longtude], zoom_start=18) 
    folium.Marker(location=[latitude, longtude], popup=name).add_to(map) 
    return map
    #map.save("map_skytree.html")
    #map.show()
```
